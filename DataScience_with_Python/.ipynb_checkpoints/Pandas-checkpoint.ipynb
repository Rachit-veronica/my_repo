### Series

```python
import numpy as np
```

```python
import pandas as pd
```

```python
labels = ['a','b','c']
my_data = [10,20,30]
arr = np.array(my_data)
d = {'a':10,'b':20,'c':30}
```

```python
# by series it add the index also and specify the data
pd.Series(my_data)
```

```python
# by specifying index we can get the index of the data
# as long as we are specifying it in correct order we didn't need to specify "data ="
pd.Series(data=my_data,index=labels)
```

```python
pd.Series(data= labels, index=my_data)     
```

```python
# and we get the same result if we pass the numpy array , work same as passing the list
pd.Series(arr)
```

```python
# if we pass any dictinary to pandas so it will do automatically assign key to index and values to corresponding-
# -data point 
pd.Series(d)
```

```python
# padas can hold any object 
# we can pass the inbuilt function also
pd.Series(data =[print,sum,max])
```

```python
 ser1=pd.Series([1,2,4,5],['USA','AUS','GER','ECU'])
```
